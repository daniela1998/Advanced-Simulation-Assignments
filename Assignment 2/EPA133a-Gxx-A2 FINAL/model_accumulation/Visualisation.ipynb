```python

csv_files = ["scenario0_summary.csv", "scenario1_summary.csv", "scenario2_summary.csv", "scenario3_summary.csv","scenario4_summary.csv", "scenario5_summary.csv", "scenario6_summary.csv",
             "scenario7_summary.csv", "scenario8_summary.csv"]


data_list = []


for file in csv_files:
    df = pd.read_csv(file)


    data_list.append((df['Average_driving_time']/60).tolist())


plt.figure(figsize=(12, 6))
plt.boxplot(data_list, patch_artist=True)


plt.xticks(range(1, 10), [f"Scenario {i}" for i in range(9)], rotation=45)
plt.title("Boxplot of All Scenarios")
plt.ylabel("Average driving time (hours)")

# Show the plot
plt.show()
```
