```python
import pandas as pd
```

```python
# import data
dir = '../infrastructure/original/'
filename = 'BMMS_overview.xlsx'
df = pd.read_excel(dir + filename)
```

```python
df.describe()
```

```python
# check for missing values
df.isnull().sum()
```

```python
# check data types
df.dtypes
```

```python
# check for duplicates on column 'LRPName' and 'road'
df.duplicated(subset=['LRPName', 'road']).sum()
```

```python
# check for unique values
df.nunique()
```

```python
# combine check for missing values, data types, duplicates, and unique values in a dataframe
data_quality_report = pd.DataFrame({'missing_values': df.isnull().sum(),
                                    'data_types': df.dtypes,
                                    'duplicates': df.duplicated().sum(),
                                    'unique_values': df.nunique()})
data_quality_report
```
